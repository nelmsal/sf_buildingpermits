```python
import pandas as pd

path = r'static_data\HPI_AT_BDL_county.xlsx'
cnty = pd.read_excel(path)
```

```python
years = [1990, 2000, 2010, 2020]
cnty = cnty[cnty['Year'].isin(years)]
```

```python
import numpy as np
cnt = pd.pivot_table(
    cnty[['State', 'County', 'Year', 'HPI']].replace('.',None), 
    values='HPI', 
    index=['State', 'County'],
    columns=['Year'],
    aggfunc=np.sum
    ).dropna().reset_index()
cnt[years] = cnt[years].astype(float)
for year in [1990, 2000, 2010]:
    cnt['diff_{}'.format(year)] = cnt[2020] - cnt[year]
    cnt['pct_{}'.format(year)] = round(((cnt[2020] - cnt[year])/cnt[2020])*100,2)
```

```python
cnt[cnt['County']=='San Francisco']
```

```python
focus = 1990
cnt.sort_values(focus, ascending=False).head(10)
```
